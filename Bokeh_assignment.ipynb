```python
# creating a Bokeh plot
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

#step1
from bokeh.plotting import figure , output_file , show

#smaple data
x = [1,2,3,4,5]
y = [4,5,6,7,8]

#Output to static HTML file
output_file("bokeh_plot.html")

#create a figure with a title
p = figure(title="My Bokeh PLot")
p.xaxis.axis_label = "X axis"
p.yaxis.axis_label = "Y label"
#add a circle
p.circle(x,y)

#show the plot
show(p)
```

Question no.2 :-

In bokeh.glphs are visual markers that represent data points on plot.

Common glyphs are:
1) Circle
2) square
3) Line
4) Triangle etc

for example:


```python
x = [1,2,3,4,5]
y = [6,7,2,4,8]

plot = figure(title = "scatter plot" , x_axis_label = "x axis" , y_axis_label = "y axis")

plot.line(x , y)
output_file("scatter_plot.html")
show(plot)
```

```python
# To customise the appearance of bokeh plot..
#Create sample data
x = [1,2,3,4,5]
y = [0,9,8,7,6]

#create a Bokeh figure
plot = figure(title = "customized plot" , x_axis_label= "X axis", y_axis_label="Y axis")

#customize other attributes
plot.title.text_color = "navy"
plot.title.text_font_size = "16px"
plot.xaxis.axis_label_text_color= "green"

plot.circle(x,y , size = 10 , color ='navy' , alpha = 0.5)

output_file("custom_plot.html")

show(plot)
```
