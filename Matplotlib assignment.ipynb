Question no.2 :-

A scatter plot is a graph that shows the relationship between two variable through plotting the graph by dots or any shape you want
* for example:=


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(3)
```

```python
x = 3 + np.random.normal(0,2,50)
y = 3 + np.random.normal(0,2,len(x))
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(6,3))
plt.scatter(x,y, color='#A87EC3')
plt.xlabel("This is my X axis")
plt.ylabel("This is my Y axis")
plt.title("This is my Scatter plot graph")
plt.show()
```

Question no. 3:-

The subplot() function in the pyplot is used to create a figure and a set of subplots.

```python
x = np.array([0,1,2,3,4,5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title("Line 1")
```

```python
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title("Line 2")
```

```python
plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title("Line 3")
```

```python
plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title("Line 4")
```

Question no.4:-

A Bar plot is a graph that shows the relationship between two variable through plotting the graph by bar diagram
* for example:=

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize=(6,3))
plt.barh(company , profit)
plt.xlabel("PROFIT")
plt.ylabel("COMPANY")
plt.title("This is my Bar plot")
```

Question no.5:-

A box plot is a special type of diagram that shows the quartiles in a box and the line extending from the lowest to the highest value.
 
It is used when showing the distribution of data points across a selected measure
* for example:=

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
data = [box1 , box2]

plt.figure(figsize=(6,3))
plt.boxplot(data , labels = ['BOX1' , 'BOX2'])
plt.title("Thi is a box plot")
plt.xlabel("box")
plt.ylabel("values")
plt.show()
```
