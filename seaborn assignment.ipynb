The five plots that can be used in the Seaborn library are 
1) Scatterplot : Explore the relationship between two continuous variables.
2) Line plot: show the trend or pattern of a single variable over time or another continuous dimension.
3) Hitogram : Display the distribution of a single variable.
4) Box plot : Summarize the distribution of a variable and identify outliers.
5) Heatmap : Visualize the correlation matrix of a dataset or show the intensity of a phenomenon across two categorial variables.

```python
flight = sns.load_dataset('flights')
```

```python
flights_pivot = flight.pivot_table(index='month', columns='year', values='passengers')

sns.heatmap(flights_pivot, cmap='YlGnBu', annot=True, fmt='d', linewidths=.5)
plt.title('Passenger Count by Month and Year')
plt.show()
```
