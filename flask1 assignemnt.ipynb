Why do we use app routes:
1) App routing allows you to organize your code by by breaking it into smaller, manageable pieces.each route corresponds to a specific functionality or page within your web applications.

2) It provides a structured and intuitive URL scheme for your application.

3) It helps maintain the separation of concerns by isolating different parts of your applications logic.
